```python
import numpy as np
import pandas as pd
import os
```

```python
home = os.path.expanduser("~")
```

```python
name = 'full_data_file.txt'  
tempdata = 'data/'+str(name)
```

```python
data = np.genfromtxt(tempdata)
```

```python
len(data[data[:,81]>40])
```

```python
data_less_40 = data[data[:,81]<40]
```

```python
len(data_less_40)
```

```python
df = pd.read_csv(home+'/Desktop/trial_runs/alllc_full_info_w_smoothing.CSV')
```

```python
red_chi_2_1 = np.abs(2*df['chi_2_1'])
red_chi_2_2 = np.abs(2*df['chi_2_2'])
```

```python
df['chi_sub']= (red_chi_2_1-red_chi_2_2)
```

```python
df_less_40 = df[((red_chi_2_1-red_chi_2_2)<40)]
```

```python
check = [[],[],[]]
for j in range(len(data_less_40)):
    n = str("cassan_%d_%d_%d.det.lc.gz" % (data_less_40[j,1],data_less_40[j,2],data_less_40[j,0]))
    if n in np.asarray(df_less_40['name']):
        check[0].append(n)
        check[1].append(1)
        check[2].append(df_less_40['chi_sub'][df_less_40['name']==n])
        continue
    else:
        check[0].append(n)
        check[1].append(0)
        check[2].append(df_less_40['chi_sub'][df_less_40['name']==n])
```

```python
df_check = pd.DataFrame({'name_matt':check[0],'is_in_mine':check[1],'chi_sub':check[2]})
```
